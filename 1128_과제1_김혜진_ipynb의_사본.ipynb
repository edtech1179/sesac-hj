```python
import pandas as pd
file_path = '/content/drive/MyDrive/01. CCTV_in_Seoul.csv'
df1 = pd.read_csv(file_path)
print(df1.head())
```

```python
file_path='/content/drive/MyDrive/01.population_in_Seoul.xls'
df2=pd.read_excel(file_path)
print(df2.head())
```

```python
population_df1 =df1[['기관명','소계']]
print(population_df1)
```

```python
print(df2.columns)
```

```python
population_df2 = df2[['자치구','인구','인구.6','65세이상고령자']]
print(population_df2)
```

```python
df1.rename(columns={'기관명':'자치구',
                        '소계':'CCTV수'
},inplace=True)

print(df1)
```

```python
selected_columns=['자치구','CCTV수']
df_with_cctv=df1[selected_columns].copy()
df_with_cctv
```

```python
import pandas as pd
df_merged = pd.merge(df_with_cctv, population_df2, on='자치구', how='inner')
df_merged
```

```python
df_merged.rename(columns={'인구.6':'외국인수'},inplace=True)
df_merged.head()
```

```python
df_merged["전체 인구 대비 CCTV비율"] = (df_merged['CCTV수']/df_merged['인구'])*100
# df_merged["외국인 수 대비  CCTV비율"] = (df_merged['CCTV수']/df_merged['외국인수'])*100
# df_merged["65세이상고령자 대비  CCTV비율"] = (df_merged['CCTV수']/df_merged['65세이상고령자'])*100
```

```python
df_merged
```

```python
df_merged["외국인 수 대비  CCTV비율"] = (df_merged['CCTV수']/df_merged['외국인수'])*100
# df_merged["65세이상고령자 대비  CCTV비율"] = (df_merged['CCTV수']/df_merged['65세이상고령자'])*100
df_merged
```

```python
df_merged["65세이상고령자 대비  CCTV비율"] = (df_merged['CCTV수']/df_merged['65세이상고령자'])*100
df_merged
```

```python
df_merged.columns
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.bar(df_merged["자치구"], df_merged["전체 인구 대비 CCTV비율"])
plt.xticks(rotation=60)
plt.ylabel("전체 인구 대비 CCTV비율")
plt.title("전체 인구 대비 CCTV비율")
plt.show()
```
