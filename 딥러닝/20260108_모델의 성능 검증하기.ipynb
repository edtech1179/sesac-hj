```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

#모델을 설정합니다
model = Sequential()
model.add(Dense(24, input_dim =60, activation = 'relu'))
model.add(Dense(10, activation = 'relu'))
model.add(Dense(10, activation = 'tanh'))
model.add(Dense(1, activation = 'sigmoid'))

#모델을 컴파일(모델 준비)합니다. : 학습 실행 전 학습 방법과 규칙 것을 모델에게 알려주는 절차
#loss: 오차 함수(채점 기준표로 모델이 예측한 답과 정답 사이 차이 계산 방법 식), optimizer: 최적화 함수(오차를 줄이기 위해 가중치를 어떻게 업데이트하는지), metrics: 평가 지표
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])

#모델을 학습합니다.
model.fit(X, Y, epochs = 200, batch_size = 10)
```

### 2. 학습셋과 테스트셋

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split

import pandas as pd
```

```python
df = pd.read_csv("C:\강의 교안\딥러닝\sonar3.csv", header=None)
```

```python
X = df.iloc[:,0:60]
Y = df.iloc[:,60]
```

```python
# 입력 받은 X와 Y를 각각 학습용과 테스트용으로 분리할 뿐, X와 Y가 무엇인지는 위에 1단계에서 직접 정해줘야 함. 
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, shuffle=True)
```

```python
model = Sequential()
model.add(Dense(24, input_dim=60, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer ='adam', metrics=['accuracy'])

history = model.fit(X_train, Y_train, epochs=200, batch_size=10)
```

```python
#모델을 테스트 셋에 적용해 정확도를 구합니다.
score = model.evaluate(X_test, Y_test)
print('Test accuracy:', score[1])
```
